# Matplotlib Complete Beginner Tutorial
This notebook will teach you Matplotlib from scratch, including line plots, bar plots, scatter plots, histograms, pie charts, and subplots.

## Step 1: Importing Matplotlib
We first need to import the library to use its functions.

```python
import matplotlib.pyplot as plt  # Importing pyplot module with alias 'plt'
```

## Step 2: Line Plot Example
A simple line plot with customizations.

```python
x = [1, 2, 3, 4, 5]  # X-axis values
y = [2, 4, 6, 8, 10]  # Y-axis values

plt.plot(x, y, color='blue', marker='o', linestyle='--', linewidth=2, label='2x Line')  # Plot line
plt.title('Simple Line Plot')  # Plot title
plt.xlabel('X-axis')  # X-axis label
plt.ylabel('Y-axis')  # Y-axis label
plt.grid(True)  # Show grid
plt.legend()  # Show legend
plt.show()  # Display the plot
```

## Step 3: Multiple Lines in One Plot
Plot two lines on the same graph for comparison.

```python
x = [0, 1, 2, 3, 4]
y1 = [0, 2, 4, 6, 8]
y2 = [0, 1, 2, 3, 4]

plt.plot(x, y1, label='2x Line', color='red', marker='o')
plt.plot(x, y2, label='y=x Line', color='green', marker='s')
plt.title('Multiple Lines in One Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

## Step 4: Bar Plot
Vertical and horizontal bar plots with labels and colors.

```python
categories = ['Apple', 'Banana', 'Cherry', 'Date']
sales = [50, 30, 70, 45]

# Vertical bar plot
plt.bar(categories, sales, color='skyblue')
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()

# Horizontal bar plot
plt.barh(categories, sales, color='orange')
plt.title('Fruit Sales (Horizontal)')
plt.xlabel('Sales')
plt.ylabel('Fruits')
plt.show()
```

## Step 5: Scatter Plot
Visualize relationships between two variables with marker size and color.

```python
x = [5, 7, 8, 7, 2, 17, 2, 9]
y = [99, 86, 87, 88, 100, 86, 103, 87]
sizes = [20, 50, 80, 200, 100, 50, 40, 60]  # Marker sizes
colors = [1, 2, 3, 4, 5, 6, 7, 8]  # Color intensity

plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7)  # Scatter plot
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Color Scale')  # Show color scale
plt.show()
```

## Step 6: Pie Chart
Show percentage distribution of categories.

```python
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [30, 25, 25, 20]
explode = (0.1, 0, 0, 0)  # Explode the first slice

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Fruit Distribution Pie Chart')
plt.show()
```

## Step 7: Histogram
Show frequency distribution of data with customizable bins.

```python
data = [22, 87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5, 79, 31, 27]
plt.hist(data, bins=5, color='purple', edgecolor='black')  # Histogram with 5 bins
plt.title('Histogram Example')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.show()
```

## Step 8: Subplots
Multiple plots in one figure using subplot.

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]

plt.figure(figsize=(10, 4))  # Set figure size

plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
plt.plot(x, y1, color='red', marker='o')
plt.title('y = x^2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

plt.subplot(1, 2, 2)  # Second plot
plt.plot(x, y2, color='blue', marker='s')
plt.title('y = (5-x)^2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

plt.tight_layout()  # Adjust spacing
plt.show()
```
